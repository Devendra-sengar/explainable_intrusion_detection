```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import glob
import os

# -----------------------
# Step 0: Load all CSVs
# -----------------------
data_path = "archive (30)"  # path to your folder with 30 CSVs
all_files = glob.glob(os.path.join(data_path, "*.csv"))

df_list = []
for f in all_files:
    df_list.append(pd.read_csv(f, low_memory=False))
df = pd.concat(df_list, ignore_index=True)
print("Dataset shape:", df.shape)
print("Columns:", df.columns)

# -----------------------
# Step 1: Minimal cleaning
# -----------------------
df = df.replace([float('inf'), -float('inf')], 0)
df = df.dropna().reset_index(drop=True)

# -----------------------
# Step 2: Encode labels
# -----------------------
le = LabelEncoder()
df['_label_multiclass'] = le.fit_transform(df['Attack Type'].astype(str))
df['_label_binary'] = (df['Attack Type'].astype(str) != 'Normal Traffic').astype(int)

print("Unique classes (multiclass):", le.classes_)
print("Binary classes:", df['_label_binary'].unique())

# -----------------------
# Step 3: Plot distributions
# -----------------------

# Multiclass
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='_label_multiclass', order=df['_label_multiclass'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Multiclass Attack Type Distribution (CICIDS2017)")
plt.xlabel("Attack Class (encoded)")
plt.ylabel("Number of Samples")
plt.tight_layout()
plt.show()

# Binary
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='_label_binary')
plt.xticks([0,1], ['Normal Traffic', 'Attack'])
plt.title("Binary Class Distribution (Normal vs Attack)")
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import glob, os

data_path = "archive (30)"  # folder containing 30 CSVs
all_files = glob.glob(os.path.join(data_path, "*.csv"))

df_list = []
for f in all_files:
    df_list.append(pd.read_csv(f, low_memory=False))

df = pd.concat(df_list, ignore_index=True)
print("Dataset shape:", df.shape)
```

```python
import numpy as np

# Replace infinities
df.replace([np.inf, -np.inf], 0, inplace=True)

# Drop missing values
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

# Encode labels
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['_label_multiclass'] = le.fit_transform(df['Attack Type'])
df['_label_binary'] = (df['Attack Type'] != 'Normal Traffic').astype(int)

print(df['_label_multiclass'].nunique(), "classes")
print(df['_label_binary'].value_counts())
```

```python
print(df.columns)
print(df['Attack Type'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------
# Step 1: Multiclass Distribution
# ------------------------
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='_label_multiclass', order=df['_label_multiclass'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Multiclass Attack Type Distribution (CICIDS2017)")
plt.xlabel("Attack Class (Encoded)")
plt.ylabel("Number of Samples")
plt.tight_layout()
plt.show()

# Optional: show counts per class as table
multiclass_counts = df['_label_multiclass'].value_counts().sort_index()
for i, count in enumerate(multiclass_counts):
    print(f"{le.classes_[i]}: {count:,} samples")

# ------------------------
# Step 2: Binary Distribution (Normal vs Attack)
# ------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='_label_binary')
plt.xticks([0,1], ['Normal Traffic', 'Attack'])
plt.title("Binary Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.tight_layout()
plt.show()

print(df['_label_binary'].value_counts())

# ------------------------
# Step 3: Feature Insights (Optional)
# ------------------------
# Top 5 numeric features overview
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_features = [f for f in numeric_features if f not in ['_label_multiclass', '_label_binary']]

df[numeric_features[:5]].describe()
```
