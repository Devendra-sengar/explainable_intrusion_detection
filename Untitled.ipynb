```python
import pandas as pd
from sklearn.utils import resample

# Suppose your full dataset is 'df'
# Desired number of samples per class
samples_per_class = 5000  # adjust based on memory

dfs = []
for cls in df['_label_multiclass'].unique():
    df_cls = df[df['_label_multiclass'] == cls]
    if len(df_cls) > samples_per_class:
        df_cls_sampled = resample(df_cls, n_samples=samples_per_class, random_state=42)
    else:
        df_cls_sampled = df_cls  # keep all if already small
    dfs.append(df_cls_sampled)

# Combine into one smaller dataset
df_small = pd.concat(dfs).sample(frac=1, random_state=42).reset_index(drop=True)

print("Small dataset shape:", df_small.shape)
print(df_small['_label_multiclass'].value_counts())
```

```python
# --------------------------
# 1 Import libraries
# --------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb
import shap

# --------------------------
# 2 Load Dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
print("Dataset shape:", df.shape)

# --------------------------
# 3 Basic EDA
# --------------------------
print("\nClass distribution:")
print(df['_label_multiclass'].value_counts())

sns.countplot(x=df['_label_multiclass'], palette="viridis")
plt.title("Class Distribution")
plt.show()

# --------------------------
# 4 Handle missing values & object columns
# --------------------------
# Drop non-numeric columns (if any) except label
df = df.select_dtypes(include=[np.number, 'int64', 'float64'])
print("Numeric features shape:", df.shape)

# Fill any remaining NaNs with median
df.fillna(df.median(), inplace=True)

# --------------------------
# 5 Split features & labels
# --------------------------
X = df.drop(columns=['_label_multiclass', '_label_binary'], errors='ignore')
y = df['_label_multiclass']

# --------------------------
# 6 Train/Test split
# --------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

# --------------------------
# 7 Hybrid Rule + XGBoost
# --------------------------
# Example rule: if Flow Duration > threshold, classify as DoS (label 3)
def hybrid_predict(X):
    preds = []
    for _, row in X.iterrows():
        # Rule-based example
        if row['Flow Duration'] > 5e7:  
            preds.append(3)  # DoS
        else:
            preds.append(np.nan)  # To be predicted by model
    return preds

# Apply rule
rule_preds = hybrid_predict(X_train)

# Prepare XGBoost training on samples not predicted by rule
mask = pd.isna(rule_preds)
X_train_model = X_train[mask]
y_train_model = y_train[mask]

# Train XGBoost
model = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=len(y.unique()),
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42
)
model.fit(X_train_model, y_train_model)

# Hybrid prediction function
def hybrid_predict_safe(X):
    preds = []
    for _, row in X.iterrows():
        if row['Flow Duration'] > 5e7:
            preds.append(3)
        else:
            r = model.predict(row.values.reshape(1, -1))[0]
            preds.append(int(r))
    return np.array(preds)

hybrid_preds = hybrid_predict_safe(X_test)

# --------------------------
# 8 Evaluation
# --------------------------
class_names = ['Bots', 'Brute Force', 'DDoS', 'DoS', 'Normal Traffic', 'Port Scanning', 'Web Attacks']
print("\nHybrid Model Classification Report:")
print(classification_report(y_test, hybrid_preds, target_names=class_names))

cm = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.show()

# --------------------------
# 9 SHAP Explainability
# --------------------------
explainer = shap.Explainer(model, X_train_model)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)

# --------------------------
# 10 Adaptive Learning Example
# --------------------------
# If new batch of data arrives
def adaptive_update(new_X, new_y):
    model.fit(new_X, new_y, xgb_model=model)
    print("Model updated with new data!")

```

```python
# --------------------------
# 1 Import Libraries
# --------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import xgboost as xgb
import shap

# --------------------------
# 2 Load Dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
print("Dataset shape:", df.shape)

# --------------------------
# 3 Basic EDA
# --------------------------
print("\nClass distribution:")
print(df['_label_multiclass'].value_counts())

sns.countplot(x=df['_label_multiclass'], palette="viridis")
plt.title("Class Distribution")
plt.show()

# --------------------------
# 4 Preprocessing
# --------------------------
# Remove non-numeric columns
df = df.select_dtypes(include=[np.number, 'int64', 'float64'])
df.fillna(df.median(), inplace=True)

# Features and Labels
X = df.drop(columns=['_label_multiclass', '_label_binary'], errors='ignore')
y = df['_label_multiclass']

# --------------------------
# 5 Train/Test Split
# --------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# --------------------------
# 6 Hybrid Rule + XGBoost Model
# --------------------------
def hybrid_predict(X):
    preds = []
    for _, row in X.iterrows():
        # Example simple rule
        if row['Flow Duration'] > 5e7:
            preds.append(3)  # DoS
        else:
            preds.append(np.nan)
    return preds

rule_preds = hybrid_predict(X_train)
mask = pd.isna(rule_preds)
X_train_model = X_train[mask]
y_train_model = y_train[mask]

# Train XGBoost
model = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=len(y.unique()),
    eval_metric='mlogloss',
    random_state=42
)
model.fit(X_train_model, y_train_model)

# Hybrid Predict Function
def hybrid_predict_safe(X):
    preds = []
    for _, row in X.iterrows():
        if row['Flow Duration'] > 5e7:
            preds.append(3)
        else:
            pred = model.predict(row.values.reshape(1, -1))[0]
            preds.append(int(pred))
    return np.array(preds)

# --------------------------
# 7 Evaluate Base Model
# --------------------------
hybrid_preds = hybrid_predict_safe(X_test)

class_names = ['Bots', 'Brute Force', 'DDoS', 'DoS', 'Normal Traffic', 'Port Scanning', 'Web Attacks']
report = classification_report(y_test, hybrid_preds, target_names=class_names, output_dict=True)
print("\nHybrid Model Classification Report:")
print(classification_report(y_test, hybrid_preds, target_names=class_names))

cm = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.show()

# --------------------------
# 8 SHAP Explainability
# --------------------------
explainer = shap.Explainer(model, X_train_model)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)

# --------------------------
# 9 Adaptive Learning System
# --------------------------
def adaptive_learning(model, new_data, new_labels):
    """
    Adaptive retraining: updates model on new samples.
    """
    print("\n Adaptive Learning Triggered...")
    model.fit(new_data, new_labels, xgb_model=model)  # continues training
    print(" Model successfully updated with new data!")

# Example: simulate new data arrival
new_samples = X_test.sample(500, random_state=99)
new_labels = y_test.loc[new_samples.index]

# Adaptive retraining
adaptive_learning(model, new_samples, new_labels)

# Re-evaluate after adaptation
hybrid_preds_updated = hybrid_predict_safe(X_test)
print("\n Post-Adaptive Learning Performance:")
print(classification_report(y_test, hybrid_preds_updated, target_names=class_names))

# Compute overall metrics
acc = accuracy_score(y_test, hybrid_preds_updated)
prec = precision_score(y_test, hybrid_preds_updated, average='weighted')
rec = recall_score(y_test, hybrid_preds_updated, average='weighted')
f1 = f1_score(y_test, hybrid_preds_updated, average='weighted')

print(f"""
 Hybrid + Adaptive Model Performance:
 Accuracy  : {acc:.4f}
 Precision : {prec:.4f}
 Recall    : {rec:.4f}
 F1-Score  : {f1:.4f}
""")
```

```python
# Improved Hybrid + XGBoost + SHAP + Adaptive Retrain pipeline
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import xgboost as xgb
import shap

# --------------------------
# Helpers
# --------------------------
def debug(msg):
    print(f"[DEBUG] {msg}")

def memory_check():
    mem = psutil.virtual_memory()
    print(f" Memory usage: {mem.percent}% ({mem.used/1e9:.2f}GB / {mem.total/1e9:.2f}GB)")

# --------------------------
# 1 Load dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
print("Dataset shape:", df.shape)

# --------------------------
# 2 Basic EDA
# --------------------------
if '_label_multiclass' not in df.columns:
    raise ValueError("Expected column '_label_multiclass' not found in dataset.")

print("\nClass distribution:")
print(df['_label_multiclass'].value_counts())

plt.figure(figsize=(8,4))
sns.countplot(x=df['_label_multiclass'], palette="viridis")
plt.title("Class Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --------------------------
# 3 Preprocessing (numeric-only, fill na)
# --------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if '_label_multiclass' not in numeric_cols:
    try:
        df['_label_multiclass'] = pd.to_numeric(df['_label_multiclass'])
        numeric_cols.append('_label_multiclass')
    except:
        raise ValueError("Could not coerce '_label_multiclass' to numeric.")

df = df[numeric_cols].copy()
df.fillna(df.median(), inplace=True)
print("After numeric filter, shape:", df.shape)

# --------------------------
# 4 Features & Target
# --------------------------
X = df.drop(columns=['_label_multiclass'], errors='ignore')
y = df['_label_multiclass'].astype(int)

unique_labels = sorted(y.unique())
print("Unique labels:", unique_labels)

# --------------------------
# 5 Train/Test Split
# --------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
memory_check()

# --------------------------
# 6 Hybrid rule definition (vectorized)
# --------------------------
THRESH_FLOW_DURATION = 5e7
FLOW_COL = 'Flow Duration'

if FLOW_COL not in X_train.columns:
    raise ValueError(f"Column '{FLOW_COL}' not found!")

rule_mask_train = X_train[FLOW_COL].values > THRESH_FLOW_DURATION
print(f"Rule triggered on {rule_mask_train.sum()} / {len(rule_mask_train)} train samples")

X_train_model = X_train.loc[~rule_mask_train].copy()
y_train_model = y_train.loc[~rule_mask_train].copy()

print("XGBoost training on", X_train_model.shape[0], "samples")

# --------------------------
# 7 Train XGBoost
# --------------------------
num_classes = len(unique_labels)
xgb_params = dict(
    objective='multi:softmax',
    num_class=num_classes,
    eval_metric='mlogloss',
    random_state=42,
    use_label_encoder=False,
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
)

model = xgb.XGBClassifier(**xgb_params)
model.fit(X_train_model.values, y_train_model.values, verbose=False)
print("XGBoost trained.")

# --------------------------
# 8 Hybrid predict (vectorized)
# --------------------------
def hybrid_predict_safe(X_df, model, flow_col=FLOW_COL, thresh=THRESH_FLOW_DURATION, rule_label=3):
    X = X_df
    mask_rule = X[flow_col].values > thresh
    preds = np.full(len(X), fill_value=-1, dtype=int)

    if mask_rule.any():
        preds[mask_rule] = int(rule_label)

    if (~mask_rule).any():
        X_for_model = X.loc[~mask_rule].values
        preds_model = model.predict(X_for_model)
        preds[~mask_rule] = preds_model.astype(int)

    return preds

# --------------------------
# 9 Evaluate before adaptation
# --------------------------
hybrid_preds = hybrid_predict_safe(X_test, model)
class_names = ['Bots', 'Brute Force', 'DDoS', 'DoS', 'Normal Traffic', 'Port Scanning', 'Web Attacks']

print("\nHybrid Model Classification Report (before adapt):")
print(classification_report(y_test, hybrid_preds, digits=4, zero_division=0, target_names=class_names))

cm = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix (before adapt)")
plt.tight_layout()
plt.show()

# --------------------------
#  SHAP Explainability
# --------------------------
print("\nComputing SHAP values...")

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test.values)

if isinstance(shap_values, list):
    mean_abs = [np.abs(s).mean() for s in shap_values]
    top_class_idx = int(np.argmax(mean_abs))
    print("Displaying SHAP summary for class index:", top_class_idx)
    shap.summary_plot(shap_values[top_class_idx], X_test, feature_names=X_test.columns)
else:
    shap.summary_plot(shap_values, X_test, feature_names=X_test.columns)

# --------------------------
#  Adaptive Learning
# --------------------------
def adaptive_learning(model, new_data, new_labels):
    if len(new_data) == 0:
        print("No new data for adaptive retraining.")
        return model

    print("\n Adaptive Learning Triggered... fitting on", len(new_data), "samples")
    try:
        model.fit(new_data.values, new_labels.values, xgb_model=model.get_booster(), verbose=False)
        print(" Model successfully updated!")
    except Exception as e:
        print(" Adaptive retrain failed:", e)

    return model

n_new = min(500, len(X_test))
new_samples = X_test.sample(n_new, random_state=99)
new_labels = y_test.loc[new_samples.index]

model = adaptive_learning(model, new_samples, new_labels)

# --------------------------
# 10 Evaluate after adaptation
# --------------------------
hybrid_preds_updated = hybrid_predict_safe(X_test, model)

print("\n Post-Adaptive Learning Performance:")
print(classification_report(y_test, hybrid_preds_updated, digits=4, zero_division=0, target_names=class_names))

cm2 = confusion_matrix(y_test, hybrid_preds_updated)
plt.figure(figsize=(10,8))
sns.heatmap(cm2, annot=True, fmt="d", cmap="Greens", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix (after adapt)")
plt.tight_layout()
plt.show()

# --------------------------
# 11 Metrics
# --------------------------
acc = accuracy_score(y_test, hybrid_preds_updated)
prec = precision_score(y_test, hybrid_preds_updated, average='weighted', zero_division=0)
rec = recall_score(y_test, hybrid_preds_updated, average='weighted', zero_division=0)
f1 = f1_score(y_test, hybrid_preds_updated, average='weighted', zero_division=0)

print(f"""
 Hybrid + Adaptive Model Performance:
 Accuracy  : {acc:.4f}
 Precision : {prec:.4f}
 Recall    : {rec:.4f}
 F1-Score  : {f1:.4f}
""")

memory_check()

# --------------------------------------------------------
# 12 SAVE MODEL  (Added as you requested)
# --------------------------------------------------------
import joblib
import os

os.makedirs("saved_model", exist_ok=True)

# Save booster (best)
model.get_booster().save_model("saved_model/hybrid_xgboost_model.json")

# Save full sklearn wrapper
joblib.dump(model, "saved_model/hybrid_xgboost_wrapper.pkl")

print("\n Model Saved Successfully:")
print(" - saved_model/hybrid_xgboost_model.json")
print(" - saved_model/hybrid_xgboost_wrapper.pkl")
```

```python
"""
hybrid_xgb_pipeline.py

Hybrid XGBoost + Rule pipeline with adaptive retraining and SHAP explainability
"""

import os
import pandas as pd
import numpy as np
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# --------------------------
# Config
# --------------------------
DATA_CSV = "CICIDS2017_small_balanced.csv"
LABEL_COL = "_label_multiclass"
FLOW_COL = "Flow Duration"
FLOW_THRESHOLD = 5e7
RULE_LABEL = 3  # DoS label
TEST_SIZE = 0.3
RANDOM_STATE = 42
N_ESTIMATORS = 200
MAX_DEPTH = 6
ADAPTIVE_COUNT = 200
SAVE_MODEL = True
MODEL_PKL = "hybrid_xgb_wrapper.pkl"

CLASS_NAMES = ['Bots','Brute Force','DDoS','DoS','Normal Traffic','Port Scanning','Web Attacks']

# --------------------------
# Helpers
# --------------------------
def hybrid_predict(X_df, model, flow_col=FLOW_COL, thresh=FLOW_THRESHOLD, rule_label=RULE_LABEL):
    """Hybrid prediction: rule first, then ML"""
    mask_rule = X_df[flow_col] > thresh
    preds = np.full(len(X_df), -1, dtype=int)
    preds[mask_rule] = rule_label
    if (~mask_rule).any():
        preds_model = model.predict(X_df.loc[~mask_rule].values)
        preds[~mask_rule] = preds_model.astype(int)
    return preds, mask_rule

def show_sample_table(samples_df, preds, mask, y_true=None):
    """Show predicted vs true table for samples"""
    print("\nSample Results:")
    table = []
    for i, row in samples_df.iterrows():
        pred_label = CLASS_NAMES[preds[i]] if preds[i]<len(CLASS_NAMES) else preds[i]
        true_label = CLASS_NAMES[y_true[i]] if y_true is not None else "N/A"
        table.append([i+1, int(row[FLOW_COL]), bool(mask[i]), pred_label, true_label])
    df_table = pd.DataFrame(table, columns=["#","Flow","Rule Fired","Predicted","True"])
    print(df_table)

def plot_class_distribution(y, title="Class Distribution"):
    plt.figure(figsize=(10,6))
    sns.countplot(x=y, palette="Set2")
    plt.title(title)
    plt.xlabel("Class Label")
    plt.ylabel("Count")
    plt.show()

# --------------------------
# Main
# --------------------------
def main():
    # 1 Load dataset
    if not os.path.exists(DATA_CSV):
        raise FileNotFoundError(f"{DATA_CSV} not found")
    df = pd.read_csv(DATA_CSV)
    df = df.select_dtypes(include=[np.number])
    df.fillna(df.median(), inplace=True)
    if LABEL_COL not in df.columns:
        raise ValueError(f"Label column {LABEL_COL} missing")
    X = df.drop(columns=[LABEL_COL])
    y = df[LABEL_COL].astype(int)

    print(f"Dataset loaded: {len(X)} rows, {X.shape[1]} features")
    plot_class_distribution(y)

    # 2 Split train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, stratify=y, random_state=RANDOM_STATE
    )

    # 3 Train XGBoost classifier
    model = xgb.XGBClassifier(
        n_estimators=N_ESTIMATORS,
        max_depth=MAX_DEPTH,
        objective="multi:softmax",
        num_class=len(np.unique(y)),
        eval_metric="mlogloss",
        use_label_encoder=False,
        verbosity=1
    )
    model.fit(X_train, y_train)
    print("\nTraining complete.")

    # 4 Test 20 synthetic/new samples
    new_samples_dicts = [
        {FLOW_COL: f, "Total Fwd Packets": np.random.randint(1,50)} for f in [
            1_000_000,2_000_000,3_000_000,800_000,400_000,5_000_000,1_500_000,600_000,50_000_001,700_000,
            900_000,1_200_000,2_200_000,3_300_000,4_400_000,800_001,1_100_000,2_500_000,6_000_000,1_300_000
        ]
    ]
    new_samples_df = pd.DataFrame(new_samples_dicts)
    # Fill missing columns with training median
    med = X_train.median()
    for c in X.columns:
        if c not in new_samples_df.columns:
            new_samples_df[c] = med[c]
    new_samples_df = new_samples_df[X.columns]

    # 5 Hybrid prediction BEFORE adaptive retrain
    preds_before, mask_before = hybrid_predict(new_samples_df, model)
    print("\n=== Predictions BEFORE Adaptive Retrain ===")
    show_sample_table(new_samples_df, preds_before, mask_before)

    # 6 Optional: SHAP explanation
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(new_samples_df)
    print("\n=== SHAP feature importance (first sample) ===")
    shap.summary_plot(shap_values, new_samples_df, plot_type="bar", show=True)

    # 7 Adaptive retrain using some misclassified / uncertain samples
    # For demo, pick random 100 samples from test set
    rng = np.random.RandomState(42)
    adaptive_idx = rng.choice(X_test.index, size=min(ADAPTIVE_COUNT, len(X_test)), replace=False)
    X_adapt = X_test.loc[adaptive_idx]
    y_adapt = y_test.loc[adaptive_idx]
    model.fit(X_adapt.values, y_adapt.values, xgb_model=model.get_booster(), verbose=True)
    print("\nAdaptive retrain complete.")

    # 8 Hybrid prediction AFTER adaptive retrain
    preds_after, mask_after = hybrid_predict(new_samples_df, model)
    print("\n=== Predictions AFTER Adaptive Retrain ===")
    show_sample_table(new_samples_df, preds_after, mask_after)

    # 9 Evaluate full test set
    preds_test, mask_test = hybrid_predict(X_test, model)
    acc = accuracy_score(y_test, preds_test)
    print(f"\nTest Accuracy after adaptive retrain: {acc:.4f}")
    plot_confusion_matrix = confusion_matrix(y_test, preds_test)
    plt.figure(figsize=(10,8))
    sns.heatmap(plot_confusion_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
    plt.title("Confusion Matrix After Adaptive Retrain")
    plt.show()

    # 10 Save trained hybrid model
    if SAVE_MODEL:
        joblib.dump(model, MODEL_PKL)
        print(f"\nHybrid model saved to {MODEL_PKL}")

if __name__ == "__main__":
    main()
```

```python
"""
full_hybrid_test_and_adapt.py

Complete testing + adaptive retrain pipeline for Hybrid Adaptive XGBoost.
(FULLY FIXED  feature name safe adaptive retraining)
"""

import os
import sys
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# --------------------------
# Configuration
# --------------------------
MODEL_PKL = "Hybrid_Adaptive_XGBoost_Model.pkl"
DATA_CSV = "CICIDS2017_small_balanced.csv"

FLOW_COL = "Flow Duration"
FLOW_THRESHOLD = 5e7
RULE_LABEL = 3

ADAPTIVE_SAMPLE_COUNT = 500

SAVE_UPDATED_MODEL = True
UPDATED_MODEL_PKL = "Hybrid_Adaptive_XGBoost_Model_updated.pkl"

CLASS_NAMES = [
    'Bots',
    'Brute Force',
    'DDoS',
    'DoS',
    'Normal',
    'PortScan',
    'WebAttack'
]

# --------------------------
# Helper Functions
# --------------------------
def check_paths():
    for path in [MODEL_PKL, DATA_CSV]:
        if not os.path.exists(path):
            print(f" Missing required file: {path}")
            sys.exit(1)


def load_model():
    print(f" Loading model: {MODEL_PKL}")
    model = joblib.load(MODEL_PKL)
    print(" Model loaded successfully")
    return model


def align_features(X_df, model):
    """
    Ensure input features match training-time features exactly
    """
    if hasattr(model, "feature_names_in_"):
        expected = list(model.feature_names_in_)
        for col in expected:
            if col not in X_df.columns:
                X_df[col] = 0
        X_df = X_df[expected]
    return X_df


def hybrid_predict(X_df, model):
    """
    Rule-based + ML hybrid prediction
    """
    mask_rule = X_df[FLOW_COL] > FLOW_THRESHOLD
    preds = np.full(len(X_df), -1, dtype=int)

    # Rule-based override
    preds[mask_rule] = RULE_LABEL

    # ML prediction for remaining samples
    if (~mask_rule).any():
        preds_ml = model.predict(X_df.loc[~mask_rule])
        preds[~mask_rule] = preds_ml.astype(int)

    return preds, mask_rule


def metrics_report(y_true, y_pred):
    print("\n Classification Report:")
    print(
        classification_report(
            y_true,
            y_pred,
            target_names=CLASS_NAMES,
            digits=4,
            zero_division=0
        )
    )

    print("Summary Metrics:")
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, average="weighted", zero_division=0))
    print("Recall   :", recall_score(y_true, y_pred, average="weighted", zero_division=0))
    print("F1 Score :", f1_score(y_true, y_pred, average="weighted", zero_division=0))


def plot_class_distribution(y):
    counts = y.value_counts().sort_index()
    plt.figure(figsize=(6, 4))
    sns.barplot(x=counts.index, y=counts.values)
    plt.title("Class Distribution")
    plt.xlabel("Class")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()


def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=CLASS_NAMES,
        yticklabels=CLASS_NAMES,
        annot_kws={"size": 9}
    )
    plt.title(title, fontsize=12)
    plt.xlabel("Predicted", fontsize=10)
    plt.ylabel("True", fontsize=10)
    plt.xticks(rotation=45, fontsize=9)
    plt.yticks(rotation=0, fontsize=9)
    plt.tight_layout()
    plt.show()


# --------------------------
#  FIXED ADAPTIVE RETRAIN
# --------------------------
def adaptive_retrain(model, X_new, y_new):
    """
    Safe warm-start retraining with feature-name preservation
    """
    if len(X_new) == 0:
        return model

    #  Enforce identical feature ordering
    if hasattr(model, "feature_names_in_"):
        X_new = X_new[model.feature_names_in_]

    print(f"\n Adaptive retraining on {len(X_new)} samples...")

    #  DO NOT use .values  preserves feature names
    model.fit(
        X_new,
        y_new,
        xgb_model=model.get_booster(),
        verbose=False
    )

    print(" Adaptive retraining complete")
    return model


# --------------------------
# Main Pipeline
# --------------------------
def main():
    check_paths()

    model = load_model()

    print(f"\n Loading dataset: {DATA_CSV}")
    df = pd.read_csv(DATA_CSV)

    # Keep only numeric features
    df = df.select_dtypes(include=[np.number])

    y = df["_label_multiclass"].astype(int)
    X = df.drop(columns=["_label_multiclass"])

    print(f"Dataset shape: {X.shape}")

    plot_class_distribution(y)

    # Align dataset to model features
    X = align_features(X, model)

    # --------------------------
    # BEFORE Adaptive Retrain
    # --------------------------
    preds_before, _ = hybrid_predict(X, model)

    print("\n Evaluation BEFORE Adaptive Retrain")
    metrics_report(y, preds_before)
    plot_confusion(y, preds_before, "Confusion Matrix (Before Adaptive)")

    # --------------------------
    # Adaptive Sampling
    # --------------------------
    rng = np.random.RandomState(777)
    idx = rng.choice(len(X), min(ADAPTIVE_SAMPLE_COUNT, len(X)), replace=False)

    X_adapt = X.iloc[idx]
    y_adapt = y.iloc[idx]

    # --------------------------
    # Adaptive Retrain
    # --------------------------
    model = adaptive_retrain(model, X_adapt, y_adapt)

    # --------------------------
    # AFTER Adaptive Retrain
    # --------------------------
    preds_after, _ = hybrid_predict(X, model)

    print("\n Evaluation AFTER Adaptive Retrain")
    metrics_report(y, preds_after)
    plot_confusion(y, preds_after, "Confusion Matrix (After Adaptive)")

    # --------------------------
    # Save Updated Model
    # --------------------------
    if SAVE_UPDATED_MODEL:
        joblib.dump(model, UPDATED_MODEL_PKL)
        print(f"\n Updated model saved as {UPDATED_MODEL_PKL}")

    print("\n Pipeline completed successfully")


if __name__ == "__main__":
    main()
```

```python
"""
full_hybrid_adaptive_production.py

FINAL PRODUCTION-GRADE Hybrid Adaptive XGBoost Pipeline
- Rule-based + ML hybrid
- Misclassification-driven adaptive retraining
- Class-focused adaptation (DDoS + Normal)
- Sample weighting for hard samples
- Feature-safe, class-safe, warm-start retraining
- Metrics and confusion matrices
"""

import os
import sys
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# =====================================================
# CONFIGURATION
# =====================================================
MODEL_PKL = "Hybrid_Adaptive_XGBoost_Model.pkl"
UPDATED_MODEL_PKL = "Hybrid_Adaptive_XGBoost_Model_updated.pkl"
DATA_CSV = "CICIDS2017_small_balanced.csv"

FLOW_COL = "Flow Duration"
FLOW_THRESHOLD = 5e7
RULE_LABEL = 3

ADAPTIVE_SAMPLE_COUNT = 2000
FOCUS_CLASSES = [2, 4]          # DDoS, Normal
ADAPTIVE_LR = 0.3
SAVE_UPDATED_MODEL = True
RANDOM_SEED = 777

CLASS_NAMES = [
    'Bots',
    'Brute Force',
    'DDoS',
    'DoS',
    'Normal',
    'PortScan',
    'WebAttack'
]

np.random.seed(RANDOM_SEED)

# =====================================================
# UTILITY FUNCTIONS
# =====================================================
def check_paths():
    for p in [MODEL_PKL, DATA_CSV]:
        if not os.path.exists(p):
            print(f" Missing file: {p}")
            sys.exit(1)

def load_model():
    print(f" Loading model: {MODEL_PKL}")
    model = joblib.load(MODEL_PKL)
    print(" Model loaded")
    return model

def align_features(X, model):
    if hasattr(model, "feature_names_in_"):
        for col in model.feature_names_in_:
            if col not in X.columns:
                X[col] = 0
        X = X[list(model.feature_names_in_)]
    return X

# =====================================================
# HYBRID PREDICTION
# =====================================================
def hybrid_predict(X, model):
    mask_rule = X[FLOW_COL] > FLOW_THRESHOLD
    preds = np.full(len(X), -1, dtype=int)
    preds[mask_rule] = RULE_LABEL
    if (~mask_rule).any():
        preds[~mask_rule] = model.predict(X.loc[~mask_rule]).astype(int)
    return preds

# =====================================================
# METRICS & PLOTTING
# =====================================================
def metrics_report(y_true, y_pred, title):
    print(f"\n {title}")
    print(classification_report(y_true, y_pred, target_names=CLASS_NAMES, digits=4, zero_division=0))
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, average="weighted", zero_division=0))
    print("Recall   :", recall_score(y_true, y_pred, average="weighted", zero_division=0))
    print("F1 Score :", f1_score(y_true, y_pred, average="weighted", zero_division=0))

def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

# =====================================================
# ADVANCED ADAPTIVE RETRAINING
# =====================================================
def adaptive_retrain(model, X, y, preds_before):
    """
    Production-ready adaptive learning:
    - Misclassified samples
    - Focus-class emphasis
    - Sample weighting
    - Feature-safe + class-safe
    - Warm-start booster
    """

    print("\n Selecting adaptive samples...")
    mis_idx = np.where(preds_before != y.values)[0]
    focus_idx = np.where(y.isin(FOCUS_CLASSES))[0]
    adapt_idx = np.unique(np.concatenate([mis_idx, focus_idx]))

    if len(adapt_idx) == 0:
        print(" No adaptive samples found")
        return model

    rng = np.random.RandomState(RANDOM_SEED)
    adapt_idx = rng.choice(adapt_idx, min(ADAPTIVE_SAMPLE_COUNT, len(adapt_idx)), replace=False)

    X_adapt = X.iloc[adapt_idx].copy()
    y_adapt = y.iloc[adapt_idx].copy()

    if hasattr(model, "feature_names_in_"):
        X_adapt = X_adapt[model.feature_names_in_]

    # Sample weighting
    weights = np.ones(len(y_adapt))
    weights[y_adapt.isin(FOCUS_CLASSES)] = 3.0
    weights[preds_before[adapt_idx] != y_adapt.values] *= 4.0

    #  CLASS-SAFE FIX
    expected_classes = model.classes_
    present_classes = np.unique(y_adapt.values)
    missing_classes = np.setdiff1d(expected_classes, present_classes)

    if len(missing_classes) > 0:
        print(f" Injecting missing classes: {missing_classes}")
        for cls in missing_classes:
            dummy_X = pd.DataFrame(np.zeros((1, X_adapt.shape[1])), columns=X_adapt.columns)
            X_adapt = pd.concat([X_adapt, dummy_X], ignore_index=True)
            y_adapt = pd.concat([y_adapt, pd.Series([cls])], ignore_index=True)
            weights = np.append(weights, 1e-6)

    print(f" Adaptive retraining on {len(X_adapt)} samples (class-safe)")

    model.set_params(learning_rate=ADAPTIVE_LR)
    model.fit(
        X_adapt,
        y_adapt,
        sample_weight=weights,
        xgb_model=model.get_booster(),
        verbose=False
    )

    print(" Adaptive retraining complete")
    return model

# =====================================================
# MAIN PIPELINE
# =====================================================
def main():
    check_paths()
    model = load_model()

    print(f"\n Loading dataset: {DATA_CSV}")
    df = pd.read_csv(DATA_CSV).select_dtypes(include=[np.number])
    y = df["_label_multiclass"].astype(int)
    X = df.drop(columns=["_label_multiclass"])
    print(f"Dataset shape: {X.shape}")

    X = align_features(X, model)

    # BEFORE adaptation
    preds_before = hybrid_predict(X, model)
    metrics_report(y, preds_before, "BEFORE Adaptive Retrain")
    plot_confusion(y, preds_before, "Confusion Matrix (Before Adaptive)")

    # ADAPTIVE LEARNING
    model = adaptive_retrain(model, X, y, preds_before)

    # AFTER adaptation
    preds_after = hybrid_predict(X, model)
    metrics_report(y, preds_after, "AFTER Adaptive Retrain")
    plot_confusion(y, preds_after, "Confusion Matrix (After Adaptive)")

    if SAVE_UPDATED_MODEL:
        joblib.dump(model, UPDATED_MODEL_PKL)
        print(f"\n Updated model saved: {UPDATED_MODEL_PKL}")

    print("\n FINAL ADVANCED PIPELINE COMPLETED SUCCESSFULLY")

# =====================================================
if __name__ == "__main__":
    main()
```
