```python
import pandas as pd
from sklearn.utils import resample

# Suppose your full dataset is 'df'
# Desired number of samples per class
samples_per_class = 5000  # adjust based on memory

dfs = []
for cls in df['_label_multiclass'].unique():
    df_cls = df[df['_label_multiclass'] == cls]
    if len(df_cls) > samples_per_class:
        df_cls_sampled = resample(df_cls, n_samples=samples_per_class, random_state=42)
    else:
        df_cls_sampled = df_cls  # keep all if already small
    dfs.append(df_cls_sampled)

# Combine into one smaller dataset
df_small = pd.concat(dfs).sample(frac=1, random_state=42).reset_index(drop=True)

print("Small dataset shape:", df_small.shape)
print(df_small['_label_multiclass'].value_counts())
```

```python
# --------------------------
# 1 Import libraries
# --------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb
import shap

# --------------------------
# 2 Load Dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
print("Dataset shape:", df.shape)

# --------------------------
# 3 Basic EDA
# --------------------------
print("\nClass distribution:")
print(df['_label_multiclass'].value_counts())

sns.countplot(x=df['_label_multiclass'], palette="viridis")
plt.title("Class Distribution")
plt.show()

# --------------------------
# 4 Handle missing values & object columns
# --------------------------
# Drop non-numeric columns (if any) except label
df = df.select_dtypes(include=[np.number, 'int64', 'float64'])
print("Numeric features shape:", df.shape)

# Fill any remaining NaNs with median
df.fillna(df.median(), inplace=True)

# --------------------------
# 5 Split features & labels
# --------------------------
X = df.drop(columns=['_label_multiclass', '_label_binary'], errors='ignore')
y = df['_label_multiclass']

# --------------------------
# 6 Train/Test split
# --------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

# --------------------------
# 7 Hybrid Rule + XGBoost
# --------------------------
# Example rule: if Flow Duration > threshold, classify as DoS (label 3)
def hybrid_predict(X):
    preds = []
    for _, row in X.iterrows():
        # Rule-based example
        if row['Flow Duration'] > 5e7:  
            preds.append(3)  # DoS
        else:
            preds.append(np.nan)  # To be predicted by model
    return preds

# Apply rule
rule_preds = hybrid_predict(X_train)

# Prepare XGBoost training on samples not predicted by rule
mask = pd.isna(rule_preds)
X_train_model = X_train[mask]
y_train_model = y_train[mask]

# Train XGBoost
model = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=len(y.unique()),
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42
)
model.fit(X_train_model, y_train_model)

# Hybrid prediction function
def hybrid_predict_safe(X):
    preds = []
    for _, row in X.iterrows():
        if row['Flow Duration'] > 5e7:
            preds.append(3)
        else:
            r = model.predict(row.values.reshape(1, -1))[0]
            preds.append(int(r))
    return np.array(preds)

hybrid_preds = hybrid_predict_safe(X_test)

# --------------------------
# 8 Evaluation
# --------------------------
class_names = ['Bots', 'Brute Force', 'DDoS', 'DoS', 'Normal Traffic', 'Port Scanning', 'Web Attacks']
print("\nHybrid Model Classification Report:")
print(classification_report(y_test, hybrid_preds, target_names=class_names))

cm = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.show()

# --------------------------
# 9 SHAP Explainability
# --------------------------
explainer = shap.Explainer(model, X_train_model)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)

# --------------------------
# 10 Adaptive Learning Example
# --------------------------
# If new batch of data arrives
def adaptive_update(new_X, new_y):
    model.fit(new_X, new_y, xgb_model=model)
    print("Model updated with new data!")

```

```python
# --------------------------
# 1 Import Libraries
# --------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import xgboost as xgb
import shap

# --------------------------
# 2 Load Dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
print("Dataset shape:", df.shape)

# --------------------------
# 3 Basic EDA
# --------------------------
print("\nClass distribution:")
print(df['_label_multiclass'].value_counts())

sns.countplot(x=df['_label_multiclass'], palette="viridis")
plt.title("Class Distribution")
plt.show()

# --------------------------
# 4 Preprocessing
# --------------------------
# Remove non-numeric columns
df = df.select_dtypes(include=[np.number, 'int64', 'float64'])
df.fillna(df.median(), inplace=True)

# Features and Labels
X = df.drop(columns=['_label_multiclass', '_label_binary'], errors='ignore')
y = df['_label_multiclass']

# --------------------------
# 5 Train/Test Split
# --------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# --------------------------
# 6 Hybrid Rule + XGBoost Model
# --------------------------
def hybrid_predict(X):
    preds = []
    for _, row in X.iterrows():
        # Example simple rule
        if row['Flow Duration'] > 5e7:
            preds.append(3)  # DoS
        else:
            preds.append(np.nan)
    return preds

rule_preds = hybrid_predict(X_train)
mask = pd.isna(rule_preds)
X_train_model = X_train[mask]
y_train_model = y_train[mask]

# Train XGBoost
model = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=len(y.unique()),
    eval_metric='mlogloss',
    random_state=42
)
model.fit(X_train_model, y_train_model)

# Hybrid Predict Function
def hybrid_predict_safe(X):
    preds = []
    for _, row in X.iterrows():
        if row['Flow Duration'] > 5e7:
            preds.append(3)
        else:
            pred = model.predict(row.values.reshape(1, -1))[0]
            preds.append(int(pred))
    return np.array(preds)

# --------------------------
# 7 Evaluate Base Model
# --------------------------
hybrid_preds = hybrid_predict_safe(X_test)

class_names = ['Bots', 'Brute Force', 'DDoS', 'DoS', 'Normal Traffic', 'Port Scanning', 'Web Attacks']
report = classification_report(y_test, hybrid_preds, target_names=class_names, output_dict=True)
print("\nHybrid Model Classification Report:")
print(classification_report(y_test, hybrid_preds, target_names=class_names))

cm = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.show()

# --------------------------
# 8 SHAP Explainability
# --------------------------
explainer = shap.Explainer(model, X_train_model)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)

# --------------------------
# 9 Adaptive Learning System
# --------------------------
def adaptive_learning(model, new_data, new_labels):
    """
    Adaptive retraining: updates model on new samples.
    """
    print("\n Adaptive Learning Triggered...")
    model.fit(new_data, new_labels, xgb_model=model)  # continues training
    print(" Model successfully updated with new data!")

# Example: simulate new data arrival
new_samples = X_test.sample(500, random_state=99)
new_labels = y_test.loc[new_samples.index]

# Adaptive retraining
adaptive_learning(model, new_samples, new_labels)

# Re-evaluate after adaptation
hybrid_preds_updated = hybrid_predict_safe(X_test)
print("\n Post-Adaptive Learning Performance:")
print(classification_report(y_test, hybrid_preds_updated, target_names=class_names))

# Compute overall metrics
acc = accuracy_score(y_test, hybrid_preds_updated)
prec = precision_score(y_test, hybrid_preds_updated, average='weighted')
rec = recall_score(y_test, hybrid_preds_updated, average='weighted')
f1 = f1_score(y_test, hybrid_preds_updated, average='weighted')

print(f"""
 Hybrid + Adaptive Model Performance:
 Accuracy  : {acc:.4f}
 Precision : {prec:.4f}
 Recall    : {rec:.4f}
 F1-Score  : {f1:.4f}
""")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_auc_score
)
from sklearn.preprocessing import label_binarize
import shap
import time

# ===========================
# 1 Evaluate Base Model
# ===========================
print(" Evaluating Base Hybrid Model...")
y_train_pred = model.predict(X_train_model)
y_test_pred = hybrid_preds

train_acc = accuracy_score(y_train_model, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"\n Train Accuracy: {train_acc:.4f}")
print(f" Test Accuracy:  {test_acc:.4f}")
print("\n Classification Report:\n", classification_report(y_test, y_test_pred, digits=4))

# ===========================
# 2 Confusion Matrix
# ===========================
plt.figure(figsize=(8,6))
cm = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix - Hybrid Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ===========================
# 3 ROCAUC Curve
# ===========================
y_test_bin = label_binarize(y_test, classes=np.unique(y))
try:
    y_proba = model.predict_proba(X_test)
    auc_score = roc_auc_score(y_test_bin, y_proba, multi_class='ovr')
    print(f"\n ROCAUC Score: {auc_score:.4f}")
except Exception as e:
    print(" Skipping AUC (model may not support predict_proba):", e)

# ===========================
# 4 Feature Importance (XGBoost + SHAP)
# ===========================
print("\n Feature Importance (Top 10):")
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=False).head(10)
print(importance_df)

# Plot Feature Importance
plt.figure(figsize=(8,5))
sns.barplot(data=importance_df, x='Importance', y='Feature', palette='viridis')
plt.title("Top 10 Important Features")
plt.show()

# ===========================
# 5 Adaptive Learning Verification
# ===========================
print("\n Testing Adaptive Learning...")

sample_X = X_train.iloc[:1000]
sample_y = y_train.iloc[:1000]

acc_before = accuracy_score(y_test, y_test_pred)
adaptive_update(sample_X, sample_y)
after_preds = hybrid_predict_safe(X_test)
acc_after = accuracy_score(y_test, after_preds)

print(f"Accuracy Before Adaptive Learning: {acc_before:.4f}")
print(f"Accuracy After  Adaptive Learning: {acc_after:.4f}")

# ===========================
# 6 Summary Report
# ===========================
summary = {
    "Train Accuracy": [train_acc],
    "Test Accuracy": [test_acc],
    "ROCAUC": [auc_score if 'auc_score' in locals() else None],
    "Adaptive Gain": [acc_after - acc_before],
    "Timestamp": [time.strftime("%Y-%m-%d %H:%M:%S")]
}

report_df = pd.DataFrame(summary)
report_df.to_csv("Hybrid_Model_Validation_Report.csv", index=False)
print("\n Report saved as 'Hybrid_Model_Validation_Report.csv'")

# ===========================
# 7 SHAP Explainability (Optional)
# ===========================
print("\n Generating SHAP Summary Plot...")
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, show=True)

print("\n Final Validation Completed Successfully!")
```

```python
# --------------------------
# 1 Import Libraries
# --------------------------
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import xgboost as xgb
import pickle
import seaborn as sns
import matplotlib.pyplot as plt

# --------------------------
# 2 Load Dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
df = df.select_dtypes(include=[np.number, 'int64', 'float64'])
df.fillna(df.median(), inplace=True)

X = df.drop(columns=['_label_multiclass', '_label_binary'], errors='ignore')
y = df['_label_multiclass']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# --------------------------
# 3 Load Saved Model
# --------------------------
with open('Hybrid_Adaptive_XGBoost_Model.pkl', 'rb') as f:
    model = pickle.load(f)
print(" Model loaded successfully!")

# --------------------------
# 4 Hybrid Prediction Function
# --------------------------
def hybrid_predict_safe(X):
    preds = []
    for _, row in X.iterrows():
        if row['Flow Duration'] > 5e7:
            preds.append(3)  # DoS
        else:
            pred = model.predict(row.values.reshape(1, -1))[0]
            preds.append(int(pred))
    return np.array(preds)

# --------------------------
# 5 Test full test set BEFORE adaptive learning
# --------------------------
hybrid_preds_test = hybrid_predict_safe(X_test)
acc_before = accuracy_score(y_test, hybrid_preds_test)
print("\n Test Accuracy BEFORE Adaptive Learning:", acc_before)
print(classification_report(y_test, hybrid_preds_test))

# --------------------------
# 6 Identify misclassified samples
# --------------------------
misclassified_indices = X_test.index[hybrid_preds_test != y_test]

# Select up to 500 misclassified samples
adaptive_batch_size = min(500, len(misclassified_indices))
X_adaptive = X_test.loc[misclassified_indices].sample(adaptive_batch_size, random_state=42)
y_adaptive = y_test.loc[X_adaptive.index]

# Make sure all classes are represented to avoid XGBoost errors
for cls in range(7):
    if cls not in y_adaptive.values:
        idx = X_test[y_test==cls].sample(1, random_state=cls*10).index[0]
        X_adaptive.loc[idx] = X_test.loc[idx]
        y_adaptive.loc[idx] = y_test.loc[idx]

print(f"\n Selected {len(X_adaptive)} samples for adaptive learning (balanced)")

# --------------------------
# 7 Predict BEFORE adaptive learning on adaptive batch
# --------------------------
preds_before = hybrid_predict_safe(X_adaptive)
print("\n Predictions BEFORE Adaptive Learning (first 20 samples)")
for i, idx in enumerate(X_adaptive.index[:20]):
    print(f"Sample {i+1}: True={y_adaptive[idx]}, Predicted={preds_before[i]}")

# --------------------------
# 8 Apply Adaptive Learning
# --------------------------
print("\n Adaptive Learning Triggered...")
model.fit(X_adaptive, y_adaptive, xgb_model=model)
print(" Model updated with adaptive samples!")

# --------------------------
# 9 Predict AFTER adaptive learning on same batch
# --------------------------
preds_after = hybrid_predict_safe(X_adaptive)
print("\n Predictions AFTER Adaptive Learning (first 20 samples)")
for i, idx in enumerate(X_adaptive.index[:20]):
    print(f"Sample {i+1}: True={y_adaptive[idx]}, Predicted={preds_after[i]}")

# --------------------------
#  Show improvements
# --------------------------
print("\n Improvements (misclassified  corrected):")
improved_count = 0
for i in range(len(X_adaptive)):
    if preds_before[i] != y_adaptive.iloc[i] and preds_after[i] == y_adaptive.iloc[i]:
        print(f"Sample {i+1}: Misclassified before  Correct after")
        improved_count += 1
if improved_count == 0:
    print("No corrections yet  adaptive batch may need more diversity.")
else:
    print(f"\nTotal improvements: {improved_count} samples corrected.")

# --------------------------
# 11 Test full set AGAIN after adaptive learning
# --------------------------
hybrid_preds_test_updated = hybrid_predict_safe(X_test)
acc_after = accuracy_score(y_test, hybrid_preds_test_updated)
print(f"\n Test Accuracy AFTER Adaptive Learning: {acc_after:.4f}")
print(classification_report(y_test, hybrid_preds_test_updated))

# --------------------------
# 12 Optional: Confusion matrix for tricky classes (DDoS=2, DoS=3)
# --------------------------
cm_before = confusion_matrix(y_test, hybrid_preds_test, labels=[2,3])
cm_after = confusion_matrix(y_test, hybrid_preds_test_updated, labels=[2,3])

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.heatmap(cm_before, annot=True, fmt="d", cmap="Reds")
plt.title("Confusion Matrix BEFORE Adaptive Learning\n(DDoS vs DoS)")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.subplot(1,2,2)
sns.heatmap(cm_after, annot=True, fmt="d", cmap="Greens")
plt.title("Confusion Matrix AFTER Adaptive Learning\n(DDoS vs DoS)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Improved Hybrid + XGBoost + SHAP + Adaptive Retrain pipeline
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import xgboost as xgb
import shap

# --------------------------
# Helpers
# --------------------------
def debug(msg):
    print(f"[DEBUG] {msg}")

def memory_check():
    mem = psutil.virtual_memory()
    print(f" Memory usage: {mem.percent}% ({mem.used/1e9:.2f}GB / {mem.total/1e9:.2f}GB)")

# --------------------------
# 1 Load dataset
# --------------------------
df = pd.read_csv("CICIDS2017_small_balanced.csv")
print("Dataset shape:", df.shape)

# --------------------------
# 2 Basic EDA
# --------------------------
if '_label_multiclass' not in df.columns:
    raise ValueError("Expected column '_label_multiclass' not found in dataset.")

print("\nClass distribution:")
print(df['_label_multiclass'].value_counts())

plt.figure(figsize=(8,4))
sns.countplot(x=df['_label_multiclass'], palette="viridis")
plt.title("Class Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --------------------------
# 3 Preprocessing (numeric-only, fill na)
# --------------------------
# Keep numeric columns only (but ensure label stays)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Ensure labels exist (they might be numeric already)
if '_label_multiclass' not in numeric_cols:
    # keep label even if non-numeric (rare); try convert
    try:
        df['_label_multiclass'] = pd.to_numeric(df['_label_multiclass'])
        numeric_cols.append('_label_multiclass')
    except:
        raise ValueError("Could not coerce '_label_multiclass' to numeric. Please convert labels first.")

df = df[numeric_cols].copy()
df.fillna(df.median(), inplace=True)
print("After numeric filter, shape:", df.shape)

# --------------------------
# 4 Features & Target
# --------------------------
# If both multiclass and binary labels present, drop binary
X = df.drop(columns=['_label_multiclass'], errors='ignore')
y = df['_label_multiclass'].astype(int)

# Save class name mapping (optional)  user provided list is safer
unique_labels = sorted(y.unique())
print("Unique labels:", unique_labels)

# --------------------------
# 5 Train/Test Split
# --------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
memory_check()

# --------------------------
# 6 Hybrid rule definition (vectorized)
# --------------------------
THRESH_FLOW_DURATION = 5e7
FLOW_COL = 'Flow Duration'  # <-- update if your column name differs

if FLOW_COL not in X_train.columns:
    raise ValueError(f"Expected flow column '{FLOW_COL}' not found. Update FLOW_COL variable accordingly.")

# rule_mask_train = True where rule triggers (we'll **not** train XGBoost on these)
rule_mask_train = X_train[FLOW_COL].values > THRESH_FLOW_DURATION
print(f"Rule triggered on {rule_mask_train.sum()} / {len(rule_mask_train)} train samples")

# Training data for XGBoost: only where rule DID NOT trigger
X_train_model = X_train.loc[~rule_mask_train].copy()
y_train_model = y_train.loc[~rule_mask_train].copy()

print("XGBoost training on", X_train_model.shape[0], "samples")

# --------------------------
# 7 Train XGBoost (multiclass)
# --------------------------
num_classes = len(unique_labels)
xgb_params = dict(
    objective='multi:softmax',
    num_class=num_classes,
    eval_metric='mlogloss',
    random_state=42,
    use_label_encoder=False,
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
)

model = xgb.XGBClassifier(**xgb_params)
model.fit(X_train_model.values, y_train_model.values, verbose=False)
print("XGBoost trained.")

# --------------------------
# 8 Hybrid predict (vectorized, fast)
# --------------------------
def hybrid_predict_safe(X_df, model, flow_col=FLOW_COL, thresh=THRESH_FLOW_DURATION, rule_label=3):
    """
    Vectorized hybrid predict:
      - if Flow Duration > thresh => rule_label (e.g. DoS)
      - else => model.predict
    Returns numpy array of predictions (same index/order as X_df)
    """
    X = X_df
    mask_rule = X[flow_col].values > thresh
    preds = np.full(len(X), fill_value=-1, dtype=int)

    if mask_rule.any():
        preds[mask_rule] = int(rule_label)

    # Predict for the rest in batch
    if (~mask_rule).any():
        X_for_model = X.loc[~mask_rule].values
        preds_model = model.predict(X_for_model)
        preds[~mask_rule] = preds_model.astype(int)

    return preds

# --------------------------
# 9 Evaluate before adaptation
# --------------------------
hybrid_preds = hybrid_predict_safe(X_test, model)
class_names = ['Bots', 'Brute Force', 'DDoS', 'DoS', 'Normal Traffic', 'Port Scanning', 'Web Attacks']

print("\nHybrid Model Classification Report (before adapt):")
print(classification_report(y_test, hybrid_preds, digits=4, zero_division=0, target_names=class_names))

cm = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix (before adapt)")
plt.ylabel("True")
plt.xlabel("Predicted")
plt.tight_layout()
plt.show()

# --------------------------
#  SHAP Explainability (TreeExplainer for XGBoost)
# --------------------------
print("\nComputing SHAP values (TreeExplainer). This may take a bit...")

explainer = shap.TreeExplainer(model)
# For multiclass, shap_values is a list (one per class)
shap_values = explainer.shap_values(X_test.values)

# Show summary plot for the most important class (example: class index 0)
# You can loop over classes if needed.
if isinstance(shap_values, list):
    # pick class with largest mean(|shap|)
    mean_abs = [np.abs(s).mean() for s in shap_values]
    top_class_idx = int(np.argmax(mean_abs))
    print("Displaying SHAP summary for class index:", top_class_idx)
    shap.summary_plot(shap_values[top_class_idx], X_test, feature_names=X_test.columns)
else:
    shap.summary_plot(shap_values, X_test, feature_names=X_test.columns)

# --------------------------
#  Adaptive Learning / Retraining
# --------------------------
def adaptive_learning(model, new_data, new_labels):
    """
    Continue training XGBoost model with more samples.
    We pass previous booster to xgb via xgb_model=model.get_booster()
    """
    if len(new_data) == 0:
        print("No new data to adapt on.")
        return model

    print("\n Adaptive Learning Triggered... fitting on", len(new_data), "new samples")
    # Use model.get_booster() as seed for continued training
    try:
        model.fit(new_data.values, new_labels.values, xgb_model=model.get_booster(), verbose=False)
        print(" Model successfully updated with new data!")
    except Exception as e:
        print(" Adaptive retrain failed:", e)
    return model

# Example: sample new data from test set as incoming data (guard sample size)
n_new = min(500, len(X_test))
new_samples = X_test.sample(n_new, random_state=99)
new_labels = y_test.loc[new_samples.index]

model = adaptive_learning(model, new_samples, new_labels)

# --------------------------
# 10 Re-evaluate after adaptation
# --------------------------
hybrid_preds_updated = hybrid_predict_safe(X_test, model)

print("\n Post-Adaptive Learning Performance:")
print(classification_report(y_test, hybrid_preds_updated, digits=4, zero_division=0, target_names=class_names))

cm2 = confusion_matrix(y_test, hybrid_preds_updated)
plt.figure(figsize=(10,8))
sns.heatmap(cm2, annot=True, fmt="d", cmap="Greens", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix (after adapt)")
plt.ylabel("True")
plt.xlabel("Predicted")
plt.tight_layout()
plt.show()

# --------------------------
# 11 Overall metrics
# --------------------------
acc = accuracy_score(y_test, hybrid_preds_updated)
prec = precision_score(y_test, hybrid_preds_updated, average='weighted', zero_division=0)
rec = recall_score(y_test, hybrid_preds_updated, average='weighted', zero_division=0)
f1 = f1_score(y_test, hybrid_preds_updated, average='weighted', zero_division=0)

print(f"""
 Hybrid + Adaptive Model Performance:
 Accuracy  : {acc:.4f}
 Precision : {prec:.4f}
 Recall    : {rec:.4f}
 F1-Score  : {f1:.4f}
""")

memory_check()
```
